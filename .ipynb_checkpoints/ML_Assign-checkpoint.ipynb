```python
df_final = df.iloc[: , 11:]
features = df_final
#features.head(1)
```

```python
x_pca = features.to_numpy()
x_pca
```

```python
# StandardScaler
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(features)

scaled_data=scaler.transform(features)
scaled_data
```

```python
#PCA Component Analysis
from sklearn.decomposition import PCA
pca=PCA(n_components=2)

pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
x_pca
```

```python
#x_pca = scaled_data
```

```python
from sklearn.cluster import KMeans
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(x_pca)
    distortions.append(kmeanModel.inertia_)
plt.plot(K, distortions, 'bx-')
plt.title('The Elbow Method showing the optimal k')    
plt.xlabel('Number of clusters')
plt.ylabel('Applicants')
plt.figure(figsize=(8,8))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,n_init=50)
kmeans.fit(x_pca)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_
print(labels)
print(centroids)
#y_kmeans = kmeans.predict(x_pca)
```

```python
plt.figure(figsize=(8,5))
colors = ['r.','g.','b.','y.']
for i in range(len(x_pca)):
    #print("labels: ",labels[i])
    plt.plot(x_pca[i][0], x_pca[i][1], colors[labels[i]], markersize =10)
plt.scatter(centroids[:,0],centroids[:,1], color= 'black', marker='x',s=300, linewidths=25, zorder = 10 )
```
